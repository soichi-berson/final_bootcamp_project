### choose top 10 and extract column names

```python
column_names = []

for name in knn.columns[:10]:
    column_names.append(name)
for name in lg.columns[:10]:
    column_names.append(name)
for name in rf.columns[:10]:
    column_names.append(name)
for name in gb.columns[:10]:
    column_names.append(name)
    
column_names
```

```python
from collections import Counter

# Count the occurrences of each value
value_counts = Counter(column_names)

# Display the count of each value
for value, count in value_counts.items():
    print(f'{value}: {count}')
```

```python
variable_frequencies_dict = {}
for value, count in value_counts.items():
    
    variable_frequencies_dict[value] = count
frequency_df = pd.DataFrame(variable_frequencies_dict.items(), columns=['Variable', 'Frequency'])

frequency_df
```

```python
sorted_df = frequency_df.sort_values(by='Frequency', ascending=False)
sorted_df
```

```python
lg2 = pd.read_csv('../data/Re_Significant_Score/down_re_LG_Significance_Score.csv')
lg2
```

### create a dataframe

```python
reshaped_df = lg2.transpose().reset_index()
reshaped_df.columns = ['factor', 'value']
reshaped_df['model'] = 'Logistic2'

reshaped_df = reshaped_df[['model', 'factor', 'value']]
reshaped_df.columns = ['model', 'factor', 'value']
reshaped_df.head(2)
```

```python
reshaped_df2 = lg.transpose().reset_index()
reshaped_df2.columns = ['factor', 'value']
reshaped_df2['model'] = 'Logistic'

reshaped_df2 = reshaped_df2[['model', 'factor', 'value']]
reshaped_df2.columns = ['model', 'factor', 'value']
reshaped_df2.head(2)
```

```python
reshaped_df3 = knn.transpose().reset_index()
reshaped_df3.columns = ['factor', 'value']
reshaped_df3['model'] = 'K-Nearest Neighbors'

reshaped_df3 = reshaped_df3[['model', 'factor', 'value']]
reshaped_df3.columns = ['model', 'factor', 'value']
reshaped_df3.head(2)
```

```python
reshaped_df4 = rf.transpose().reset_index()
reshaped_df4.columns = ['factor', 'value']
reshaped_df4['model'] = 'Random Forest'

reshaped_df4 = reshaped_df4[['model', 'factor', 'value']]
reshaped_df4.columns = ['model', 'factor', 'value']
reshaped_df4.head(2)
```

```python
reshaped_df5 = gb.transpose().reset_index()
reshaped_df5.columns = ['factor', 'value']
reshaped_df5['model'] = 'Gradient Boosting'

reshaped_df5 = reshaped_df5[['model', 'factor', 'value']]
reshaped_df5.columns = ['model', 'factor', 'value']
reshaped_df5.head(2)
```

```python
# concatnate all dataframes
factors = pd.concat([reshaped_df2,reshaped_df3,reshaped_df4,reshaped_df5,reshaped_df], axis=0)
factors.reset_index(drop=True)
```

### Get the percentage of features

```python
reshaped_df2.head(2)
```

```python
LG = reshaped_df2.copy()
LG['percentage'] = (LG['value'] / LG['value'].sum() * 100).round(2)
LG.head()
```

```python
KNN = reshaped_df3.copy()
KNN['percentage'] = (KNN['value'] / KNN['value'].sum() * 100).round(2)
KNN.head()
```

```python
RF = reshaped_df4.copy()
RF['percentage'] = (RF['value'] / RF['value'].sum() * 100).round(2)
RF.head()
```

```python
GB = reshaped_df5.copy()
GB['percentage'] = (GB['value'] / GB['value'].sum() * 100).round(2)
GB.head()
```

```python
#combine

# Merge LG and KNN DataFrames
merged_df1 = pd.merge(LG[['factor', 'percentage']], KNN[['factor', 'percentage']], on='factor')

# Merge RF and GB DataFrames
merged_df2 = pd.merge(RF[['factor', 'percentage']], GB[['factor', 'percentage']], on='factor')

# Combine the merged DataFrames
merged_df = merged_df1.merge(merged_df2, on='factor')

merged_df.head()
```

```python
merged_df['percentages(AVG)'] = ((merged_df['percentage_x_x'] + merged_df['percentage_y_x'] +
merged_df['percentage_x_y'] + merged_df['percentage_y_y']) /4) .round(2)

merged_df.head()
```

```python
factors4 = merged_df[['factor', 'percentages(AVG)']]
factors4
```
